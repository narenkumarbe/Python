```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 10, 0.1)
```

```python
time
```

```python
amplitude = np.sin(time)
```

```python
amplitude
```

```python
plt.plot(time, amplitude)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```
